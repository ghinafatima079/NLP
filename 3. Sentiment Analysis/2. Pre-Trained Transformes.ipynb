# Transformers 

Transformers can understand context

instead of looking at words in isolation, the understand how meaning changes depending on the surrounding words

allows them to handle complex sentences, subtle tones and even sarcasm much more effectively than rule based analysis

```python
import transformers
from transformers import pipeline #pipeline provides a quick way to use pre-trained models for various NLP tasks without having to build and train models from scratch

```

```python
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
sentence1 = "I love programming!"
sentence2 = "I hate bugs."
sentence3 = "The movie was okay, not great but not terrible."
sentence4 = "The food was absolutely wonderful, I enjoyed every bite."
sentence5 = "It is raining today."
```

```python
for sentence in [sentence1, sentence2, sentence3, sentence4, sentence5]:
    print(sentence)
    print(sentiment_pipeline(sentence))
    print('\n')
```

the model misclassified the 5th sentence as positive instead of neutral. to improve results, use a different pre-trained model

```python
specific_model_pipeline = pipeline("sentiment-analysis", model="finiteautomata/bertweet-base-sentiment-analysis")
```

```python
for sentence in [sentence1, sentence2, sentence3, sentence4, sentence5]:
    print(sentence)
    print(specific_model_pipeline(sentence))
    print('\n')
```
