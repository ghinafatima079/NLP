lexicon : a collection of words paired with their sentiment scores

TextBlob and Vader use their own set of lexicons which is why they may produce different results when analyzing the same text

```python
sentence1 = "I love programming!"
sentence2 = "I hate bugs."
sentence3 = "The movie was okay, not great but not terrible."
sentence4 = "The food was absolutely wonderful, I enjoyed every bite."
sentence5 = "It is raining today."
```

# TextBlob

```python
from textblob import TextBlob
```

```python
print("TextBlob Sentiment Analysis:\n")
for sentence in [sentence1, sentence2, sentence3, sentence4, sentence5]:
    sentiment_score = TextBlob(sentence).sentiment.polarity
    print(f"Sentence: {sentence}")
    print(f"Sentiment Score: {sentiment_score} \n")
```

# Vader

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
vader_sentiment_analyzer = SentimentIntensityAnalyzer()
```

```python
for sentence in [sentence1, sentence2, sentence3, sentence4, sentence5]:
    sentiment_scores = vader_sentiment_analyzer.polarity_scores(sentence)
    print(f"Sentence: {sentence}")
    print(f"VADER Sentiment Scores: {sentiment_scores} \n")
```
