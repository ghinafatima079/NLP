```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer #breaks a text into words and counts the number of times each word occurs
#gives us the numerical representation of the text data we need for machine learning algorithms to work
```

```python
data = ['Most Shark attacks occur near the coast. It is important to be cautious while swimming in the ocean.',
        'The new movie received rave reviews from critics and audiences alike. It is a must-watch for everyone.',
        'The restaurant had poor service and the food was bland. I would not recommend it to anyone.',
        'The concert was an unforgettable experience. The band played all their hit songs and the crowd was energetic.',
        'The book was a thrilling page-turner from start to finish. The plot twists kept me on the edge of my seat.',
        'The hotel had a beautiful view of the beach and the staff were friendly and accommodating.',
        'The traffic was terrible and I was late for my appointment. I will avoid that route in the future.']
```

```python
countvec = CountVectorizer()
```

```python
countvec_fit = countvec.fit_transform(data) #fit: looks through the data and learns which unique words are present. Transform: converts teh text data into numerical representation and counts the occurrences of each word
```

```python
bag_of_words_df = pd.DataFrame(countvec_fit.toarray(), columns=countvec.get_feature_names_out()) #this will give us a dataframe representation of the bag of words model by converting the sparse matrix to a dense array and using the unique words as column names
#toarray() method converts the sparse matrix to a dense array
#columns=countvec.get_feature_names_out() gets the unique words learned during the fit process to use as column names
```

```python
print(bag_of_words_df)
```
