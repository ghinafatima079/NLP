# Lemmatization

reduces a token to its base form while preserving the intended meaning, using a pretrained dictionary Wordnet

```python
import nltk
nltk.download('wordnet')
```

```python

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
tokens = ["connecting", "connected", "connection", "connects", "connectable", "connectivity", "reconnecting", "worse", "studies", "studying"]

for t in tokens:
    print(t, ": ", lemmatizer.lemmatize(t))
```
