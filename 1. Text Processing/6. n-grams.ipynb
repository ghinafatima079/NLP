# N-grams

helps us analyze the relationship between neighbouring words

n-grams = a sequence of tokens

in 'i love nlp', 

n-gram = 1 : unigram(single-word tokens)  eg - ['i', 'love', 'nlp']

n-gram = 2 : bigram  eg - ['i love', 'love nlp']

n-gram = 3 : trigram  eg - ['i love nlp']

```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tokens = ['the', 'rise', 'of', 'ai', 'in', 'modern', 'technology', 'and', 'its', 'impact', 'on', 'society', 'is', 'undeniable', '.', 'from', 'healthcare', 'to', 'finance', ',', 'ai', 'is', 'transforming', 'industries', 'and', 'shaping', 'the', 'future', '.', 'however', ',', 'it', 'also', 'raises', 'ethical', 'concerns', 'regarding', 'privacy', ',', 'bias', ',', 'and', 'job', 'displacement', '.', 'as', 'we', 'navigate', 'this', 'new', 'era', ',', 'it', 'is', 'crucial', 'to', 'strike', 'a', 'balance', 'between', 'innovation', 'and', 'responsibility', '.', 'collaborative', 'efforts', 'from', 'governments', ',', 'industries', ',', 'and', 'academia', 'are', 'essential', 'to', 'ensure', 'that', 'ai', 'is', 'developed', 'and', 'deployed', 'in', 'a', 'manner', 'that', 'benefits', 'all', 'of', 'humanity', '.']

print("Total number of tokens:", len(tokens))
```

pd.Series is a single-column data in a spreadsheet. organizes our ngrams in a structured format that pandas can work with

nltk.ngrams takes two arguments - (list of tokens, value of n)

```python
unigrams = (pd.Series(nltk.ngrams(tokens, 1)).value_counts())
print(unigrams)
```

Visualizing the unigrams using a bar chart


```python
unigrams[:10].sort_values().plot.barh(color="lightsalmon", width = .9, figsize=(5,3))

#barh = horizontal bar plot
#figsize = (width, height) of the plot in inches
#sort_values() to sort the unigrams in ascending order for better visualization
#[:10] to select the top 10 unigrams

plt.title("Top 10 Unigrams")
plt.xlabel("Frequency")
plt.ylabel("Unigrams")
plt.show()
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens, 2)).value_counts())
print(bigrams)
```

```python
trigrams = (pd.Series(nltk.ngrams(tokens, 3)).value_counts())
print(trigrams)
```
