# Stemming

reducing the words to their base form to standardize and reduce complexity


drawback - output may or may not be meaningful

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

tokens = ["connecting", "connected", "connection", "connects", "connectable", "connectivity", "reconnecting", "worse", "studies", "studying"]
```

```python
for t in tokens:
    print(t, ": ", ps.stem(t))
```
