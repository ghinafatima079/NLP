## Raw String

```python
result_search = re.search(r"pattern", "The string with a pattern inside")
print(result_search)
```

```python
customer_reviews = [
    "Sarah,I love this product! It works great and exceeded my expectations.",
    "Terrible quality. Broke after one use. Do not buy!",
    "Average item. Not bad, but nothing special either.",
    "Fantastic Sarah! Highly recommend to everyone",
    "Waste of money Sara. Very disappointed with the purchase", 
] 
```

```python
#finding reviews with sarah in them

sarahs_reviews = []

pattern_to_find = r"Sarah?"

for string in customer_reviews:
    if re.search(pattern_to_find, string):
        sarahs_reviews.append(string)
        
print(sarahs_reviews)
```

```python
#finding reviews starting with A

a_reviews = []

pattern_to_find = r"^A"

for review in customer_reviews:
    if re.search(pattern_to_find, review):
        a_reviews.append(review)

print(a_reviews)
```

```python
#finding reviews ending with e

e_reviews = []

pattern_to_find = r"e$"
for review in customer_reviews:
    if re.search(pattern_to_find, review):
        e_reviews.append(review)

print(e_reviews)
        
```

```python
#finding reviews with exceeded or disappointed in them

exceeded_disappointed_reviews = []

pattern_to_find = r"(disappoint|exceed)ed"

for review in customer_reviews:
    if re.search(pattern_to_find,review):
        exceeded_disappointed_reviews.append(review)

print(exceeded_disappointed_reviews)
```

```python
#reviews without punctuations

pattern_to_find = r"[^\w\s]" #matches any character that is not a word character or whitespace

reviews_no_punctuations = []

for review in customer_reviews:
    review_no_punct = re.sub(pattern_to_find, "", review)
    reviews_no_punctuations.append(review_no_punct)
    
print(reviews_no_punctuations)
```
