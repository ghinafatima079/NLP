```python
import nltk
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
en_stopwords = stopwords.words('english')
print(en_stopwords)
```

```python
sentence1 = "This is a sample sentence, showing off the stop words filtration."
sentence2 = "It is too far to go to the shop and he did not want her to walk"
```

```python
sentence1_no_stopwords = ' '.join([word for word in sentence1.split() if word.lower() not in en_stopwords])
sentence2_no_stopwords = ' '.join([word for word in sentence2.split() if word.lower() not in en_stopwords])
print(sentence1_no_stopwords)
print(sentence2_no_stopwords)
```
